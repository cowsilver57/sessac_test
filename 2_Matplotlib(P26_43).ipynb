```python
#P27-28 (x축 세부 설정)
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(2,2))
ax.set_xlim([0,10]) #x축을 0부터 10까지의 숫자로 설정
ax.set_xticks([0,1,5,10]) #x축을 0부터 10까지 0,1,5,10의 간격으로 설정
```

```python
# (반복문으로 x축의 값 설정하기)
fig, ax = plt.subplots(figsize=(2,2))
ax.tick_params(labelsize=5) #축의 글씨 크기 설정

xticks = [i for i in range(10)] #x축을 0부터 9까지 반복해서 설정
ax.set_xticks(xticks) #여기까지 해줘야 그래프에 구현이 됨!
```

```python
# (반복문을 활용하여 X축의 범위 설정과 함께 간격도 설정하기)
fig, ax = plt.subplots(figsize=(4,4))

xticks = [i for i in range(0,101,20)] #(시작 값, 끝 값, 간격)
ax.set_xticks(xticks)
ax.tick_params(axis='x',
               labelsize=10, length=10,
               width=3, rotation=30)
```

```python
#P30 (major, minor ticks 설정하기)
fig, ax = plt.subplots(figsize=(4,2))

major_xticks = [i for i in range(0,101,20)]
minor_xticks = [i for i in range(0,101,5)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks, minor=True)

ax.tick_params(axis='x', labelsize=10, length=10, width=3, rotation=30)
ax.tick_params(axis='x', which='minor', length=5, width=2)
```

```python
#P31-32 (실습)
fig, ax = plt.subplots(figsize=(5,3))

x_major_ticks = [i for i in range(0,101,20)]
x_minor_ticks = [i for i in range(0,101,5)]

y_major_ticks = [i for i in range(0,11,2)]
y_minor_ticks = [i for i in range(0,11,1)]

ax.set_xticks(x_major_ticks)
ax.set_xticks(x_minor_ticks, minor=True)

ax.set_yticks(y_major_ticks)
ax.set_yticks(y_minor_ticks, minor=True)

ax.tick_params(axis='x', labelsize=10, length=5, width=3, rotation=30)
ax.tick_params(axis='x', which='minor', length=3, width=1)
ax.tick_params(axis='y', labelsize=10, length=5, width=3)
ax.tick_params(axis='y', which='minor', length=3, width=1)
```

```python
#P33-36 (그래프 안에 넣는 글씨 색깔 설정하기)

color_list = ['b','g','r','c','m','y']

fig, ax = plt.subplots(figsize=(3,5))
ax.set_xlim([-1,1])
ax.set_ylim([-1, len(color_list)])

#enumerate 함수: 반복 가능한(iterable) 객체(예: 리스트, 튜플, 문자열)를 순회하면서 해당 요소와 해당 요소의 인덱스를 함께 반환하기 위해 사용
#기본 형태: enumerate(iterable, start=0)
#                   iterable: 순회하고자 하는 반복 가능한 객체(리스트, 튜플, 문자열 등)를 나타냄.
#                   start (선택적): 인덱스의 시작값을 나타내며, 기본값은 0임.

# color_list에 있는 각 색상을 반복적으로 처리하기 위한 루프를 시작.
# c_idx는 현재 색상의 인덱스를 나타내고 c는 현재 색상 자체를 나타냄.
for c_idx, c in enumerate(color_list):
    ax.text(0,c_idx,
            "color="+c,
            fontsize=20,
            ha='center',
            color=c)
#이 줄은 텍스트를 그래프에 추가하는 부분입니다. 각 색상에 대한 정보를 텍스트로 표시합니다.
# 0은 x 좌표를 나타내며, c_idx는 y 좌표를 나타냄. 따라서 각 텍스트는 y 축의 다른 위치에 표시됨.
#"color="+c는 표시될 텍스트 내용이며, color=c는 텍스트의 색상을 c로 설정함.
```

```python
# (RGB를 활용하여 색깔 설정하기)
color_list = [(1.,0.,0.),
              (0.,1.,0.),
              (0.,0.,1.)]
fig, ax = plt.subplots(figsize=(3,5))
ax.set_xlim([-1,1])
ax.set_ylim([-1, len(color_list)])

for c_idx, c in enumerate(color_list):
    ax.text(0,c_idx,
            f"color={c}",
            fontsize=10, ha='center', color=c)
```

```python
#P37-38 (scatter, plot 그래프 생성하기)
import numpy as np

#seed를 설정하면 난수 생성 과정이 예상되기 때문에 random 함수를 실행할 때, seed를 설정함.
np.random.seed(0)

n_data = 100
#평균이 0이고 표준 편차가 1인 정규 분포에서 무작위로 추출된 n_data 개의 데이터 포인트를 생성
x_data = np.random.normal(0,1,(n_data))
y_data = np.random.normal(0,1,(n_data))

fig, ax = plt.subplots(figsize=(5,5))

#방법 1 -> scatter를 이용하여 바로 표현하기
ax.scatter(x_data, y_data, s=100) #(x축 기준, y축 기준, 포인트 크기=markersize)

#방법 2 -> plot을 이용하여 'o'모양의 점 찍어서 그래프 표현하기
#ax.plot(x_data, y_data, 'o', markersize=10)
```

```python
#P39 (
import numpy as np
np.random.seed(0)

x_min, x_max = -5,5
n_data = 300

#x_min과 x_max 사이의 값을 무작위로 추출하여 n_data개의 데이터 포인트 생성
x_data = np.random.uniform(x_min, x_max, n_data)
#x_data에 노이즈를 추가하기 위해 np.random.normal 함수를 사용하여 난수 생성
y_data = x_data + 0.5*np.random.normal(0,1,n_data)

#x_min과 x_max 사이를 균일한 간격으로 나눈 2개의 데이터 포인트 생성
#선형 예측을 활용하기 위해서는 linspace의 괄호 마지막 값이 2가 됨.
#두 개의 양 끝점을 이으면 직선이 도출되기 때문이라고 생각하면 편함!
pred_x = np.linspace(x_min, x_max, 2)
pred_y = pred_x

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x_data, y_data)

ax.plot(pred_x, pred_y, color='r', linewidth=3)
```

```python
#P40(scatter plot에서 size array)

n_data = 10
x_data = np.linspace(0,10,n_data)
y_data = np.linspace(0,10,n_data)

#값이 커질 수록 데이터 포인트의 크기도 커지게 하기 위해서 사용
s_arr = np.linspace(10,500,n_data)

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x_data, y_data, s=s_arr)
```

```python
#P41 (scatter plot에서 color array)

n_data = 10
x_data = np.linspace(0,10,n_data)
y_data = np.linspace(0,10,n_data)

c_arr = [(c/n_data, c/n_data, c/n_data) for c in range(n_data)]

fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(x_data, y_data, s=300, c=c_arr)
```

```python
#P42 (scatter plot에서 size, color array 같이 표현하기)
import numpy as np

n_data = 10
x_data = np.linspace(0,10,n_data)
y_data = np.linspace(0,10,n_data)

s_arr = np.linspace(10,500,n_data)
c_arr = [(c/n_data, c/n_data, c/n_data) for c in range(n_data)]

fig, ax = plt.subplots(figsize=(3,3))

ax.scatter(x_data, y_data, s_arr, c_arr)
```

```python
#P43 ()
n_data = 500
x_data = np.random.normal(0,1,size=(n_data,))
y_data = np.random.normal(0,1,size=(n_data,))
s_arr = np.random.uniform(100,500,n_data)
#(0,1,3)인 이유는 RGB를 표현하는 방법을 나타내기 때문임.
#P36를 참고하면 RGB를 표현할 때, (0.5, 0.4, 0.9) 이런식이기 때문!!
#이 것을 n_data만큼 반복해서 컬러 설정~!
c_arr = [np.random.uniform(0,1,3) for _ in range(n_data)]

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x_data, y_data, s=s_arr, c=c_arr, alpha=0.5)
```
