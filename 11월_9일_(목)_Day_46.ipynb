# Affine Function의 구현

##1. affine1,2의 객체를 만들고 객체들의 w,b 출력해보기

```python
class AF:
	def __init__(self, w, b):
			self.w = w
			self.b = b

affine1_obj = AF(w=[1.1],b=-1.5)
affine2_obj = AF(w=[-1.-1],b=0.5)

print("affine1\n", f'w:{affine1_obj.w}', f'b:{affine1_obj.b}')
print("affine2\n", f'w:{affine2_obj.w}', f'b:{affine2_obj.b}')
```

##2. x를 입력받아 z = w*x + b를 연산하는 기능(forward method)를 가진 AffineFunction 클래스

```python
import numpy as np
class AF:
    def __init__(self, w, b):
        self.w = w
        self.b = b
    def forward(self, x):
        z = np.dot(self.w, x) + self.b
        return z
```

# Sigmoid Function의 구현

## sigmoid 코드로 구현하고 계산해보기

```python
import numpy as np

class Sigmoid:
    def forward(self, z):
        a = 1 / (1 + np.exp(-z))
        return a

sigmoid = Sigmoid()

print("z=-5:", sigmoid.forward(z=-5))
print("z=-3:", sigmoid.forward(z=-3))
print("z=0:", sigmoid.forward(z=0))
print("z=3:", sigmoid.forward(z=3))
print("z=5:", sigmoid.forward(z=5))
```

##sigmoid 그래프 그려보기

```python
#내 답
import matplotlib.pyplot as plt

class Sigmoid:
    def forward(self, z):
        a = 1 / (1 + np.exp(-z))
        return a

sigmoid = Sigmoid()

z_values = np.linspace(-6, 6, 100)
sigmoid_values = [sigmoid.forward(z) for z in z_values]

fig, ax = plt.subplots(figsize=(10,7))
plt.plot(z_values, sigmoid_values, label='Sigmoid Function')
plt.title('Sigmoid Function')
ax.axvline(x=0, color='black', linestyle='--', linewidth=1)
ax.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.show()
```

```python
#강사님 답
import numpy as np
import matplotlib.pyplot as plt

class Sigmoid:
    def forward(self, z):
        a = 1 / (1 + np.exp(-z))
        return a

sigmoid = Sigmoid()

x = np.linspace(-5,5,100)
y = sigmoid.forward(z=x)

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x,y)

ax.axhline(y=0, color='gray', linewidth=1)
ax.axvline(x=0, color='gray', linewidth=1)

ax.tick_params(labelsize=15)
plt.show()
```

# Artificial Neuron 구현

```python
#내 답
import numpy as np

class AF:
    def __init__(self,w,b):
        self.w = w
        self.b = b

    def forward(self,x):
        z = np.dot(self.w,x) + self.b
        return z

class SF:
    def forward(self,z):
        a = 1 / (1+np.exp(-z))
        return a

class AN:
    def __init__(self, affine_w, affine_b):
        self.affine = AF(affine_w, affine_b)
        self.activation = SF()

    def forward(self, x):
        z = self.affine.forward(x)
        a = self.activation.forward(z)
        return a
```
