```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#P54~55 ax.axvline & ax.axhline (수직선, 수평선)
fig, ax = plt.subplots(figsize=(5,5))

ax.set_xlim([-5,5])
ax.set_ylim([-5,5])

ax.axvline(x=1, color='r', linewidth=3)
ax.axvline(x=2, ymax=0.8, ymin=0.2, color='b', linewidth=3)

ax.axhline(y=1, color='r', linewidth=2)
ax.axhline(y=2, xmax=0.8, xmin=0.2, color='b', linewidth=2)
```

```python
#P56 그래프에 수직선과 수평선 추가하기
x = np.linspace(-4*np.pi, 4*np.pi, 200)
sin = np.sin(x)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,sin)
#ls = line style
ax.axhline(y=1, ls=':', lw=1, color='gray')
ax.axhline(y=-1, ls=':', lw=1, color='gray')
```

```python
#P57~59 Line styles
x_data = np.array([0,1])
y_data = x_data

#첫 번째 그래프
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x_data, y_data)
ax.plot(x_data, y_data+1, linestyle=':', color='r' )
ax.plot(x_data, y_data+2, linestyle='-.', color='b')
ax.plot(x_data, y_data+3, linestyle='--', color='gray')

#두 번째 그래프
t = np.linspace(-4*np.pi, 4*np.pi, 200)
fig, ax = plt.subplots(figsize=(7,5))

ax.plot(t,sin,color='black')
ax.axhline(y=1, ls=':', color='r')
ax.axhline(y=-1, ls=':', color='r')
```

```python
#P60~61 Markers
t = np.linspace(-4*np.pi, 4*np.pi, 100)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(7,5))
ax.plot(t, sin+1, marker='o', color='black',
        markersize=10, markerfacecolor='pink',
        markeredgecolor='r', markeredgewidth=3)
```

```python
#P62 Basic Usage of Legend
np.random.seed(0)

n_data = 100
random_noise1 = np.random.normal(0,1,(n_data,))
random_noise2 = np.random.normal(1,1,(n_data,))
random_noise3 = np.random.normal(2,1,(n_data,))

fig, ax = plt.subplots(figsize=(7,5))
ax.tick_params(labelsize=10)
ax.plot(random_noise1, label='random noise1')
ax.plot(random_noise2, label='random noise2')
ax.plot(random_noise3, label='random noise3')
ax.legend(fontsize=10)
```
