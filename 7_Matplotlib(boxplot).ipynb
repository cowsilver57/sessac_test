```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#P2-4 boxplot 기본 설정
#학생 100명 -> 사실 이 데이터는 없어도 아래의 그래프 생성 가능임,
n_student = 100
#평균이 50, 표준편차가 10인 100개의 랜덤 데이터 생성
math_scores = np.random.normal(loc=50, scale=10, size=(100,))

fig, ax = plt.subplots(figsize=(7,7))
ax.boxplot(math_scores, notch=True, whis=2, sym='bx', showfliers=False, vert=False)
#1. notch = True -> 중앙값의 신뢰 구간을 표시하는 함수
#2. whis는 이상치의 길이를 조절하는 함수 -> 박스에서 튀어나온 선의 길이가 조정됨.
#3. sym는 이상치의 표시 방법을 정하는 함수 -> 기본적으로 원으로 저장되어 있음.
#4. showfliers는 이상치를 표현할지 말지 정하는 함수 -> 기본적으로 False로 저장되어 있음.
#5. vert는 그래프를 수직으로 표현할지 말지 정하는 함수임.
```

```python
math_scores = np.random.normal(loc=50, scale=10, size=(100,))
fig, ax = plt.subplots(figsize=(7,7))

#P5 boxplot 중앙값선 설정 (k는 검정색을 의미함)
medianprops = {'linewidth':2, 'color':'k'}

#P6 boxplot 박스선 설정
boxprops = {'linestyle':'--', 'color':'k', 'alpha':0.7}

#P7 bocplot 이상치선 설정
whiskerprops = {'linestyle':'--', 'color':'pink','alpha':0.8}

#마지막으로 프린트하기
ax.boxplot(math_scores,
           medianprops=medianprops,
           boxprops=boxprops,
           whiskerprops=whiskerprops,
           capprops=whiskerprops) #capprops = 이상치의 최대 최소 선을 의미함.
```

```python
#P8 boxplot 실습 1
n_student = 100
data1 = np.random.normal(loc=50, scale=15, size=(100,1))
data2 = np.random.normal(loc=70, scale=10, size=(n_student,1))
data3 = np.random.normal(loc=30, scale=12, size=(n_student,1))
data4 = np.random.normal(loc=80, scale=5, size=(n_student,1))

data = np.hstack((data1, data2, data3, data4))

fig, ax = plt.subplots(figsize=(10,7))
ax.set_ylim([0,100])
ax.boxplot(data)
```

```python
#P9 boxplot 실습 2
#그래프를 예쁘게 그리기 위해 seaborn 라이브러리 사용
plt.style.use('seaborn')

#데이터 정의하기
n_student=100
math_scores = np.random.normal(loc=50, scale=15, size=(100,1))
english_scores = np.random.normal(loc=70, scale=10, size=(n_student,1))
physics_scores = np.random.normal(loc=30, scale=20, size=(n_student,1))
programming_scores = np.random.normal(loc=80, scale=5, size=(n_student,1))
#데이터 수평으로 쌓기
data = np.hstack((math_scores, english_scores, physics_scores, programming_scores))

#그래프 설정 전처리
medianprops = {'linewidth':1.5, 'color':'red'}
boxprops = {'linewidth':1.5, 'color':'k', 'alpha':0.7}
whiskerprops = {'linestyle':'--', 'color':'blue', 'alpha':0.7}

labels = ['Math','English','Physics','Programming']

#그래프 설정
fig, ax = plt.subplots(figsize=(10,7))
ax.boxplot(data,
           labels=labels,
           medianprops=medianprops,
           boxprops=boxprops,
           whiskerprops=whiskerprops,
           capprops=whiskerprops,
           notch=True)
#y축 범위 설정
ax.set_ylim([1,100])
#y축 대표 범위 설정
major_yticks = np.arange(0,101,20)
minor_yticks = np.arange(0,101,5)

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(labelsize=20)
ax.tick_params(axis='x', labelrotation=10)
ax.grid(axis='y', linewidth=2)
ax.grid(axis='y', which='minor', linewidth=2, linestyle=':')
ax.grid(axis='x', linewidth=0)
```

```python
#P10 boxplot 실습 3
#그래프를 예쁘게 그리기 위해 seaborn 라이브러리 사용
plt.style.use('seaborn')

#데이터 정의하기
n_student=100
math_scores = np.random.normal(loc=50, scale=15, size=(100,1))
english_scores = np.random.normal(loc=70, scale=10, size=(n_student,1))
physics_scores = np.random.normal(loc=30, scale=20, size=(n_student,1))
programming_scores = np.random.normal(loc=80, scale=5, size=(n_student,1))
#데이터 수평으로 쌓기
data = np.hstack((math_scores, english_scores, physics_scores, programming_scores))

#그래프 설정 전처리
medianprops = {'linewidth':1.5, 'color':'red'}
boxprops = {'linewidth':1.5, 'color':'k', 'alpha':0.7}
whiskerprops = {'linestyle':'--', 'color':'blue', 'alpha':0.7}

labels = ['Math','English','Physics','Programming']

#그래프 설정
fig, axes = plt.subplots(2,1,figsize=(10,7))
axes[0].boxplot(data)
axes[1].violinplot(data)

#위의 그래프 설정
axes[0].set_ylim([0,100])
axes[0].tick_params(labelsize=20, bottom=False, labelbottom=False)
axes[1].set_ylim([0,100])
axes[1].tick_params(labelsize=20)

#그래프 사이의 거리 설정하기
fig.subplots_adjust(hspace=0.1)
```
