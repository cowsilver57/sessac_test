```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from sklearn.datasets import make_blobs
from torch.utils.data import TensorDataset, DataLoader

N_SAMPLES = 100

X1, y1 = make_blobs(n_samples=N_SAMPLES, centers=([-1,1],[-1,-1]), n_features=2, cluster_std=0.3)
X2, y2 = make_blobs(n_samples=N_SAMPLES, centers=([1,-1], [1,1]), n_features=2, cluster_std=0.3)

X = np.concatenate((X1, X2), axis=0)
y = np.concatenate((y1, y2), axis=0)

fig, ax = plt.subplots(figsize=(10,10))

X_neg = X[y == 0]
X_pos = X[y == 1]

ax.scatter(X_neg[:,0], X_neg[:,1], color='red', label = 'label = 0')
ax.scatter(X_pos[:,0], X_pos[:,1], color='blue', label = 'label = 1')

ax.legend(fontsize=15, loc='upper left')
ax.tick_params(labelsize=15)

plt.show()
```
