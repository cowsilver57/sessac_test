```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#P63~64 Legend의 위치
n_data = 100
random_noise1 = np.random.normal(0,1,(n_data,))
random_noise2 = np.random.normal(1,1,(n_data,))
random_noise3 = np.random.normal(2,1,(n_data,))

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(random_noise1, label='random noise1')
ax.plot(random_noise2, label='random noise2')
ax.plot(random_noise3, label='random noise3')
ax.tick_params(labelsize=5)
ax.legend(fontsize=10, loc='upper right')
#가장 많이 쓰는 legend의 위치는 upper right, upper left, lower right, lower left
```

```python
#P66~67 nocl Argument -> legend를 몇 분단으로 표현할 것인지 결정하는 함수
PI = np.pi
t = np.linspace(-4*PI, 4*PI, 200)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(7,7))

for ax_idx in range(10):
    label_template = 'added by {}'
    ax.plot(t, sin+ax_idx, label=label_template.format(ax_idx))
ax.legend(fontsize=10, ncol=2)
```

```python
#P68~71 bbox_to_anchor Argument -> legend랑 비슷한 개념임
n_data = 100
random_noise1 = np.random.normal(0,1,(n_data,))
random_noise2 = np.random.normal(1,1,(n_data,))
random_noise3 = np.random.normal(2,1,(n_data,))

fig, ax = plt.subplots(figsize=(5,5))
ax.tick_params(labelsize=5)

ax.plot(random_noise1, label='random noise1')
ax.plot(random_noise2, label='random noise2')
ax.plot(random_noise3, label='random noise3')

#위치를 선정할 때, bbox를 기준으로 두고 생각해야함.
ax.legend(fontsize=10, bbox_to_anchor=(1,0.5), loc='center left')
#가장 많이 쓰는 위치로는 center right, center left, lower center, upper center 등이 있음. -> 그래프 밖에 표현 가능하기 때문임.
fig.tight_layout()
```

```python
#P72 bbox_to_anchor Argument 실습
PI = np.pi
t = np.linspace(-4*PI, 4*PI, 200)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(10,10))

for ax_idx in range(12):
    label_template = 'added by {}'
    ax.plot(t, sin+ax_idx, label=label_template.format(ax_idx))

ax.legend(fontsize=10, ncol=4, bbox_to_anchor=(0.5, -0.05), loc='upper center')

fig.tight_layout()
```

```python
#P73 Line Styles/Markers with Legend 실습
PI = np.pi
t = np.linspace(-4*PI, 4*PI, 200)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(10,10))

ax.plot(t, sin, label='sin(t)')
ax.plot(t, sin+1, marker='o', label='sin(t)+1', linestyle=':')
ax.plot(t, sin+2, marker='D', label='sin(t)+2', linestyle='--')
ax.plot(t, sin+3, marker='s', label='sin(t)+3', linestyle='-.')

ax.legend(loc='center left', bbox_to_anchor=(1,0.5), fontsize=15)
fig.tight_layout()
```
