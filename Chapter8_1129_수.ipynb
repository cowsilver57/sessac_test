```python
import pandas as pd

X_test = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/304_x_test.csv')
X_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/304_x_train.csv')
y_train = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/304_y_train.csv')

#데이터 살펴보기
# print(X_test.head())
# print(X_train.head())
# print(y_train.head())

# print(X_test.info())
# print(X_train.info())
# print(y_train.info())

# print(X_test.isnull().sum())
# print(X_train.isnull().sum())
# print(y_train.isnull().sum())

#데이터 분류하기
COL_DEL = ['ID']
COL_NUM = ['Age','AnnualIncome','FamilyMembers','ChronicDiseases']
COL_CAT = ['Employment Type','GraduateOrNot','FrequentFlyer','EverTravelledAbroad']
COL_Y = ['TravelInsurance']

#데이터 분할하기
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(X_train[COL_NUM+COL_CAT],
                                            y_train[COL_Y].values.ravel(),
                                            test_size=0.3,
                                            stratify=y_train[COL_Y].values.ravel())
# print(y_tr)

#데이터 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_tr[COL_NUM])

X_tr[COL_NUM] = scaler.transform(X_tr[COL_NUM])
X_val[COL_NUM] = scaler.transform(X_val[COL_NUM])
X_test[COL_NUM] = scaler.transform(X_test[COL_NUM])

# print(X_tr[COL_NUM].head())

#데이터 인코딩
from sklearn.preprocessing import LabelEncoder
X = pd.concat([X_train[COL_CAT], X_test[COL_CAT]])

for col in COL_CAT:
    le = LabelEncoder()
    le.fit(X[col])
    X_tr[col] = le.transform(X_tr[col])
    X_val[col] = le.transform(X_val[col])
    X_test[col] = le.transform(X_test[col])

    # print(col)
    # print(le.classes_)
    # print('\n')

#모델 학습하기
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(random_state=123)
modelRF.fit(X_tr, y_tr)

from xgboost import XGBClassifier
modelXGB = XGBClassifier(random_state=123)
modelXGB.fit(X_tr, y_tr)

#예측값 생성하기
y_val_predRF = modelRF.predict_proba(X_val)[:,1]
y_val_predXGB = modelXGB.predict_proba(X_val)[:,1]

#평가 지표 구하기
from sklearn.metrics import roc_auc_score
scoreRF = roc_auc_score(y_val, y_val_predRF)
scoreXGB = roc_auc_score(y_val, y_val_predXGB)
print('Random Forest:', scoreRF)
print('XGBoost:', scoreXGB)

#함수 생성하기
def get_scores(model, X_tr, x_val, y_tr, y_val):
    y_tr_pred = model.predict_proba(X_tr)[:,1]
    y_val_pred = model.predict_proba(X_val)[:,1]
    tr_score = roc_auc_score(y_tr, y_tr_pred)
    val_score = roc_auc_score(y_val, y_val_pred)
    return f"train: {round(tr_score,4)}, valid: {round(val_score,4)}"

#성능 확인하기
print('\n')
print('Random Forest:\n', get_scores(modelRF, X_tr, X_val, y_tr, y_val))
print('XGBoost:\n', get_scores(modelXGB, X_tr, X_val, y_tr, y_val),'\n')

#과대적합으로 RF 부적절! XGBoost 사용
#하이퍼 파라미터 조정하기
modelXGB2 = XGBClassifier(n_estimators=50, max_depth=3, min_child_weight=1, random_state=123)
modelXGB2.fit(X_tr, y_tr)

modelXGB3 = XGBClassifier(n_estimators=50, max_depth=5, min_child_weight=1, random_state=123)
modelXGB3.fit(X_tr, y_tr)

modelXGB4 = XGBClassifier(n_estimators=50, max_depth=3, min_child_weight=2, random_state=123)
modelXGB4.fit(X_tr, y_tr)

modelXGB5 = XGBClassifier(n_estimators=50, max_depth=5, min_child_weight=2, random_state=123)
modelXGB5.fit(X_tr, y_tr)

print('XGBoost2:\n', get_scores(modelXGB2, X_tr, X_val, y_tr, y_val))
print('XGBoost3:\n', get_scores(modelXGB3, X_tr, X_val, y_tr, y_val))
print('XGBoost4:\n', get_scores(modelXGB4, X_tr, X_val, y_tr, y_val))
print('XGBoost5:\n', get_scores(modelXGB5, X_tr, X_val, y_tr, y_val))
```

##1. 단일표본 T-검정

```python
import pandas as pd
import scipy.stats as stats
from math import sqrt

a = pd.read_csv('/content/yemoonsaBigdata/datasets/supplement/trees.csv')
X = a['Height'].mean()
print(round(X,2))

t_score, p_value = stats.ttest_1samp(a['Height'],75)
print(round(t_score,2))

print(round(p_value,4))
if p_value>0.05:
    print('채택')
else:
    print('기각')
```

##2. 독립표본 T-검정

```python
import pandas as pd
import scipy.stats as stats
from math import sqrt

a1 = pd.read_csv('/content/yemoonsaBigdata/datasets/supplement/toothgrowth.csv')

stats.levene(a1.loc[a1.supp=='VC','len'], a1.loc[a1.supp=='OJ','len'])

VC_a1 = a1.loc[a1.supp=='VC']
OJ_a1 = a1.loc[a1.supp=='OJ']

VC_X = VC_a1['len'].mean()
OJ_X = OJ_a1['len'].mean()

print(round(VC_X,2), round(OJ_X,2))

t_score, p_value = stats.ttest_ind(VC_a1['len'], OJ_a1['len'], equal_var=True)
print(round(t_score,2))

print(round(p_value,4))
if p_value>0.05:
    print('채택')
else:
    print('기각')
```

##3. 쌍체표본 T-검정

```python
import pandas as pd
import scipy.stats as stats
from math import sqrt

b = pd.read_csv('/content/yemoonsaBigdata/datasets/supplement/insectsprays.csv')

diff = b['after_spr']-b['before_spr']
m_d = diff.mean()
print(round(m_d,2))

t_score, p_value = stats.ttest_rel(b['before_spr'], b['after_spr'], alternative='greater')
print(round(t_score,2))

print(round(p_value,4))
if p_value>0.05:
    print('채택')
else:
    print('기각')
```

##4. 일원분산분석

```python
import pandas as pd
import scipy.stats as stats
from math import sqrt

b1 = pd.read_csv('/content/yemoonsaBigdata/datasets/supplement/iris.csv')

result = b1.loc[b1['species']=='versicolor','sepal_width'].mean()
print(round(result,2))

X1 = b1.loc[b1['species']=='setosa','sepal_width']
X2 = b1.loc[b1['species']=='versicolor','sepal_width']
X3 = b1.loc[b1['species']=='virginica', 'sepal_width']

test_result = stats.f_oneway(X1, X2, X3)
print(round(test_result.statistic,2))

p_value = test_result.pvalue
print(round(p_value,4))

if p_value>0.05:
    print('채택')
else:
    print('기각')
```

##5. 카이제곱 적합도 검정

```python
import pandas as pd
import scipy.stats as stats
from math import sqrt

c = pd.read_csv('/content/yemoonsaBigdata/datasets/supplement/cellphone.csv')

count_df = c.groupby('제조사').count()
print(count_df)

result = count_df.loc['삼성','id'] - count_df.loc['애플','id']
print(result)

total_sum = len(c)
expected = [int(total_sum*0.1),
            int(total_sum*0.6),
            int(total_sum*0.3)]
print(expected)

observed = count_df['id'].values
test_result = stats.chisquare(observed, f_exp = expected)
result = round(test_result.statistic,2)
print(result)

p_value = test_result.pvalue
print(round(p_value,4))

if p_value>0.05:
    print('채택')
else:
    print('기각')
```

##6. 카이제곱 독립성 검정

```python
import pandas as pd
import scipy.stats as stats
from math import sqrt

d = pd.read_csv('/content/yemoonsaBigdata/datasets/supplement/student.csv')
table = pd.crosstab(d['단과대학'], d['수강 과목 수'])
print(table)

result = table[table['6개 이상']==table['6개 이상'].max()].index[0]
print(result)

X1 = table.loc['경영대학',:]
X2 = table.loc['자연대학',:]
X3 = table.loc['공과대학',:]

test_result = stats.chi2_contingency([X1,X2,X3])
print(round(test_result.statistic,2))

p_value = test_result.pvalue
print(round(p_value,4))

if p_value>0.05:
    print('채택')
else:
    print('기각')
```
